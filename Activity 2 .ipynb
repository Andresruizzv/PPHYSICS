```python
import random
import matplotlib.pyplot as plt

# Configuración de la caja y partículas
width, height = 100, 80  # Dimensiones de la caja
num_particles = 20       # Número de partículas
radius = 2               # Radio de cada partícula

# Lista para almacenar las posiciones de las partículas
particles = []

def distance(p1, p2):
    """Calcula la distancia euclidiana entre dos puntos."""
    return ((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)**0.5

# Generar partículas que no se superpongan
while len(particles) < num_particles:
    x = random.uniform(radius, width - radius)
    y = random.uniform(radius, height - radius)
    new_particle = (x, y)

    # Comprobar que no se superpone con partículas existentes
    if all(distance(new_particle, p) > 2 * radius for p in particles):
        particles.append(new_particle)

# Visualización de la caja y las partículas
fig, ax = plt.subplots()
ax.set_xlim(0, width)
ax.set_ylim(0, height)
ax.set_aspect('equal')
ax.set_title('Caja con Partículas aleatorias')

# Dibujar partículas
for (x, y) in particles:
    circle = plt.Circle((x, y), radius, color='blue', fill=True)
    ax.add_patch(circle)

# Dibujar bordes de la caja
rect = plt.Rectangle((0, 0), width, height, linewidth=1, edgecolor='black', facecolor='none')
ax.add_patch(rect)

plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Configuración de la caja y partículas
width, height = 100, 80  # Dimensiones de la caja
num_particles = 20       # Número de partículas
radius = 2               # Radio de cada partícula

# ===========================
# Primera caja: distribución uniforme
# ===========================
# Calcular el espaciado para distribuir partículas uniformemente
cols = int(num_particles**0.5)  # Número de columnas (cuadrada raíz aproximada)
rows = (num_particles // cols) + (1 if num_particles % cols != 0 else 0)  # Número de filas
x_spacing = width / cols  # Espaciado entre partículas en x
y_spacing = height / rows  # Espaciado entre partículas en y

# Generar posiciones uniformemente distribuidas
uniform_particles = []
for row in range(rows):
    for col in range(cols):
        if len(uniform_particles) < num_particles:  # Asegurar que no se excedan las partículas
            x = x_spacing / 2 + col * x_spacing  # Posición x de la partícula
            y = y_spacing / 2 + row * y_spacing  # Posición y de la partícula
            uniform_particles.append((x, y))  # Guardar la posición

# ===========================
# Segunda caja: distribución aleatoria sin superposición
# ===========================
random_particles = []

def distance(p1, p2):
    """Calcula la distancia euclidiana entre dos puntos."""
    return ((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)**0.5

# Generar partículas que no se superpongan
while len(random_particles) < num_particles:
    x = random.uniform(radius, width - radius)  # Generar posición x aleatoria
    y = random.uniform(radius, height - radius)  # Generar posición y aleatoria
    new_particle = (x, y)  # Crear nueva partícula

    # Comprobar que no se superpone con partículas existentes
    if all(distance(new_particle, p) > 2 * radius for p in random_particles):
        random_particles.append(new_particle)  # Agregar partícula si no hay superposición

# ===========================
# Visualización de las dos cajas
# ===========================
fig, axs = plt.subplots(1, 2, figsize=(12, 6))  # Crear dos gráficos lado a lado

# Configuración para la primera caja (uniforme)
axs[0].set_xlim(0, width)  # Límites en x
axs[0].set_ylim(0, height)  # Límites en y
axs[0].set_aspect('equal')  # Escala uniforme
axs[0].set_title('Distribución Uniforme')  # Título del gráfico

# Dibujar partículas uniformes
for (x, y) in uniform_particles:
    circle = plt.Circle((x, y), radius, color='blue', fill=True)  # Crear círculo
    axs[0].add_patch(circle)  # Agregar círculo al gráfico

# Dibujar bordes de la caja
rect = plt.Rectangle((0, 0), width, height, linewidth=1, edgecolor='black', facecolor='none')
axs[0].add_patch(rect)  # Agregar borde de la caja

# Configuración para la segunda caja (aleatoria)
axs[1].set_xlim(0, width)  # Límites en x
axs[1].set_ylim(0, height)  # Límites en y
axs[1].set_aspect('equal')  # Escala uniforme
axs[1].set_title('Distribución Aleatoria')  # Título del gráfico

# Dibujar partículas aleatorias
for (x, y) in random_particles:
    circle = plt.Circle((x, y), radius, color='red', fill=True)  # Crear círculo
    axs[1].add_patch(circle)  # Agregar círculo al gráfico

# Dibujar bordes de la caja
rect = plt.Rectangle((0, 0), width, height, linewidth=1, edgecolor='black', facecolor='none')
axs[1].add_patch(rect)  # Agregar borde de la caja

# Mostrar el gráfico
plt.tight_layout()  # Ajustar diseño
plt.show()
```
